```python
import nltk
from nltk import CFG
from nltk.grammar import FeatureGrammar, Nonterminal
from nltk.parse import RecursiveDescentParser, FeatureEarleyChartParser
from nltk.parse.generate import generate
from collections import Counter
from nltk import FreqDist


text = open("brexit_britain_will_be_just_fine.txt").readlines()
text = [l for l in text if l.strip()]
text = " ".join(text)
print(text)
```

```python
sentences = nltk.sent_tokenize(text)
tokens = nltk.word_tokenize(text)

print(sentences)

words = [v for v in tokens if v[0].isalnum()]
words_lowercase = [v.lower() for v in words]
vocabulary = set(words_lowercase)

fdist = FreqDist(words_lowercase)
hapaxes = fdist.hapaxes()


print("The text containts", len(vocabulary), "distinct words.")
print("The 10 most common words are:", ", ".join([x[0] for x in fdist.most_common(10)]) + ".")
print("Some hapaxes are:", ", ".join(hapaxes[:10]) + ".")


fdist.plot(50)
```

```python
# complex tags
complex_pos_tag =nltk.pos_tag(vocabulary)
for tag in complex_pos_tag:
    word_type = tag[1]+ " -> "+ '\'' +tag[0] +'\''
    print(word_type)

# simple tags
# simple_pos_tag =nltk.pos_tag(vocabulary, tagset='universal')
# for tag in simple_pos_tag:
#     word_type = tag[1]+ " -> "+ '\'' +tag[0] +'\''
#     print(word_type)
       
```
